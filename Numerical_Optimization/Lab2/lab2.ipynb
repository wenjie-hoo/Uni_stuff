```python
def generate_regular_polygon_centers(K, d):
    angles = np.linspace(0, 2 * np.pi, K, endpoint=False)
    centers_x = d * np.cos(angles)
    centers_y = d * np.sin(angles)
    return centers_x, centers_y

def generate_point_clouds(K, d):
    centers_x, centers_y = generate_regular_polygon_centers(K, d)
    point_clouds = []
    for i in range(K):
        ai = centers_x[i]
        bi = centers_y[i]
        x = np.random.normal(ai, 1, 1000)
        y = np.random.normal(bi, 1, 1000)
        point_clouds.append((x, y))
    return point_clouds,centers_x, centers_y

def plot_point_clouds(K, d, point_clouds):
    plt.figure(figsize=(10, 8))
    for i in range(K):
        plt.scatter(point_clouds[i][0], point_clouds[i][1])#, label=f'Center {i+1}')
    plt.title(f'Point Clouds for K={K}, d={d}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.savefig(f'point_clouds_K_{K}_d_{d}.png')
    plt.show()
```

```python
K_values = [7, 11, 23]
d_values = [5, 10, 15]

for K in K_values:
    for d in d_values:
        point_clouds,_,_ = generate_point_clouds(K, d)
        plot_point_clouds(K, d, point_clouds)
```

```python
def compute_distances(point_clouds, centers_x, centers_y):
    distances = []
    for i in range(len(point_clouds)):
        distances_center = []
        for j in range(len(point_clouds)):
            dist = np.sqrt((point_clouds[i][0] - centers_x[j])**2 + (point_clouds[i][1] - centers_y[j])**2)
            distances_center.append(dist)
        distances.append(distances_center)
    return distances

def color_points(point_clouds, distances):
    colors = []
    red_points = 0
    for i in range(len(point_clouds)):
        color = []
        for j in range(len(point_clouds[i][0])):
            min_dist = distances[i][i][j]
            closest_cloud = i
            for k in range(len(point_clouds)):
                if k != i and distances[i][k][j] < min_dist:
                    min_dist = distances[i][k][j]
                    closest_cloud = k
            if closest_cloud == i:
                color.append('green')
            else:
                color.append('red')
                red_points += 1
        colors.append(color)
    print('percentage of red points:',red_points / len(color))
    return colors

def plot_colored_point_clouds(point_clouds, colors):
    plt.figure(figsize=(10, 8))
    for i in range(len(point_clouds)):
        plt.scatter(point_clouds[i][0], point_clouds[i][1], color=colors[i])
    plt.title('Colored Point Clouds')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()

K = 10
d = 8

point_clouds, centers_x, centers_y = generate_point_clouds(K, d)
distances = compute_distances(point_clouds, centers_x, centers_y)
colors = color_points(point_clouds, distances)
plot_colored_point_clouds(point_clouds, colors)
```

```python
def distance_matrix(X, Y):
    D = np.sqrt(np.sum(X**2, axis=0, keepdims=True).T - 
                           2 * np.dot(X.T, Y) + np.sum(Y**2, axis=0, keepdims=True))
    np.fill_diagonal(D, 0)
    return D

d = 100
X1,Y1 = np.random.rand(d, 1000),np.random.rand(d, 1000)
X2, Y2 = np.random.rand(d, 10000), np.random.rand(d, 1000)

s_t = time.time()
D1 = distance_matrix(X1, Y1)
t_1 = time.time() - s_t

s_t = time.time()
D2 = distance_matrix(X2, Y2)
t_2 = time.time() - s_t

print("1000*1000:", t_1, "sec")
print("10000*1000:", t_2, "sec")

# Print the distance matrices
print("d=100, N=1000, M=1000:")
print(D1)

print("\nd=100, N=10000, M=1000:")
print(D2)
```

```python
def nearest_neighbor(X, Y):
    distances = np.sqrt(np.sum((X[:, :, np.newaxis] - Y[:, np.newaxis, :])**2, axis=0))
    return np.array(np.argmin(distances, axis=1))

def k_nearest_neighbors(X, Y, k):
    k_nearest_ids = np.zeros((k, X.shape[1]), dtype=int)
    for i in range(X.shape[1]):
        distances = np.sqrt(np.sum((X[:, i][:, np.newaxis] - Y)**2, axis=0))
        nearest_ids = np.argsort(distances)[:k]
        k_nearest_ids[:, i] = nearest_ids
    return np.array(k_nearest_ids)

nearest_ids = nearest_neighbor(X2, Y2)
print(nearest_ids)

k_nearest_ids = k_nearest_neighbors(X2, Y2, k=5)
print(k_nearest_ids)
```
