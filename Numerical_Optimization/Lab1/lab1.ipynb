```python
data_10k = np.random.uniform(-1, 1, 10000)
data_100k = np.random.uniform(-1, 1, 100000)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_10k, bins=100, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('10k')
plt.xlabel('Value')
plt.ylabel('Density')

x = np.linspace(-1, 1, 1000)
pdf_uniform = np.ones_like(x) / 2
plt.plot(x, pdf_uniform, color='red', linestyle='--', linewidth=2, label='PDF')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(data_100k, bins=100, density=True, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title('100k')
plt.xlabel('Value')
plt.ylabel('Density')

plt.plot(x, pdf_uniform, color='red', linestyle='--', linewidth=2, label='PDF')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
points_10k = [np.random.normal(5, 2, 10000), np.random.normal(3, 1, 10000)]
points_100k = [np.random.normal(5, 2, 100000), np.random.normal(3, 1, 100000)]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(points_10k[0], points_10k[1], alpha=0.5, s=10)
plt.title('10k')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 2, 2)
plt.scatter(points_100k[0], points_100k[1], alpha=0.5, s=10)
plt.title('100k')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
