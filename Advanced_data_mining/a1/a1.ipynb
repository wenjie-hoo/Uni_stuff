```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min
from scipy.stats import norm
```

```python
k = 0.5
lambda_ = 60
# weibull_dist = weibull_min(k, scale=lambda_)

Tu_values = weibull_min.rvs(k, scale=lambda_, size=1000)
time_points = [0, 150, 24*60, 2*24*60, 5*24*60]

for t in time_points:
    active_users = np.sum(Tu_values >= t)
    print(f"At time {t} minutes, {active_users} users are active.")
    
time_list = np.arange(0, 7*24*60)

active_users = []
for t in time_list:
    active_users.append(np.sum(Tu_values >= t))
    
plt.plot(time_list, active_users)
plt.xlabel("Time (minutes)")
plt.ylabel("Number of active users")
plt.title("Number of active users over time")
plt.show()
```
