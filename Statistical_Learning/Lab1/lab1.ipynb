### Problem 2

1. Since we know the columns of $P$ are made of eigenvector $e_i, i=1...n$ are orthonormalized, it implies that $e_i^Te_i=1$ and $e_i^Te_j=0$, for $i\neq j$. When we transpose an ort matrix, the $P^T$ is still orthnomal, and its inverse $P^{-1}$ exists and is equal to $P^T$, because $PP^T=I$ where $I$ is the identity matrix. 
$$P^TP = \begin{bmatrix} e_1^T \\\vdots \\ e_n^T \end{bmatrix} \begin{bmatrix} e_1 & \cdots & e_n \end{bmatrix} = \begin{bmatrix} e_1^Te_1 & \cdots & e_1^Te_n \\ \vdots & \ddots & \vdots \\ e_n^Te_1 & \cdots & e_n^Te_n \end{bmatrix} = I$$


2. In the spectral decomposition $A=P\Lambda P^T$, we let $\Lambda=diag(\lambda_1,\lambda_2,...,\lambda_n)$, then the $\Lambda_i$ are the roots of the polynomial, then the polynomial factors as $det(A-\lambda I)=(\lambda_1-\lambda)(\lambda_2-\lambda)\cdot\cdot\cdot(\lambda_n-\lambda)$, we assume the abstract variable $\lambda=0$, we see that $det(A)=\lambda_1\lambda_2\lambda_3$, using the fact from 3 $det(A)=det(\Lambda)$, so the $det(\Lambda_n)=\prod_i^n{\lambda_i}$

3. In the spectral decompositon $A=P\Lambda P^T$, we can observe that $det(A)=det(P\Lambda P^T)=det(P)\cdot det(\Lambda)\cdot det(P^T)$, since $P$ and $P^T$ are ort matrices, the $det(P)=det(P^T)=\pm 1$, thus $det(A)=det(\Lambda)$.

4. Since $\lambda=diag(\lambda_1,\lambda_2,...,\lambda_n)$ is a diagonal matrix, then $\Lambda^{-1}=diag({1\over\lambda_1},{1\over\lambda_2},...,{1\over\lambda_n})$ 

5.  To verify that $A^-1=P\Lambda^-1P^T$, we need to show that $AA^-1=I$. Given $A=P\Lambda P^T$, we compute the product $AA^-1=P\Lambda P^T\cdot P\Lambda^{-1}P^T=P\Lambda\Lambda^{-1}P^T=PIP^T=PP^T=I$, therefore, $A^{-1}=P\Lambda^{-1}P^T$ indeed defines the inverse of A.

```r
A <- matrix(c(3, -1, -1, 3), nrow = 2, byrow = TRUE)
eigen_A <- eigen(A)
lambda <- eigen_A$values
P <- eigen_A$vectors
cat("Eigenvalues:",lambda)
cat("\nEigenvectors (columns of P):\n")
print(P)
# 2.1
cat("\n2.1")
print(all.equal(t(P), solve(P)))
# 2.2
cat("\n2.2")
print(all.equal(det(A), prod(eigen_A$values)))
# 2.3
cat("\n2.3")
print(all.equal(det(A), det(diag(eigen_A$values))))
# 2.4
cat("\n2.4")
A_inv <- P %*% diag(1/lambda) %*% t(P)
print(A_inv)
# 2.5 
cat("\n2.5")
print(all.equal(A_inv, P %*% diag(1/lambda) %*% t(P)))
```

4. How many children would score zero, one, and two, respectively? Illustrate this

```r
weight_percentiles <- quantile(data$Weight, c(0.75, 0.95))
length_percentiles <- quantile(data$Height, c(0.75, 0.95))

# Define the scoring function
determine_score <- function(weight, length) {
  if (is.na(weight) || is.na(length)) {
    return(NA)  # Return NA if either weight or length is missing
  } else if (!is.na(weight_percentiles[2]) && weight > weight_percentiles[2]) {
    return(0)
  } else if (!is.na(length_percentiles[2]) && length > length_percentiles[2]) {
    return(0)
  } else if (!is.na(weight_percentiles[1]) && weight > weight_percentiles[1]) {
    return(1)
  } else if (!is.na(length_percentiles[1]) && length > length_percentiles[1]) {
    return(1)
  } else {
    return(2)
  }
}

scores <- apply(data, 1, function(row) {
  if (is.na(row[1]) || is.na(row[2])) {
    print("NA value detected in weight or length")
    return(NA)
  }
  determine_score(row[1], row[2])
})
scores
```

```r
score_counts <- table(scores)

ggplot(data, aes(x = Weight, y = Length, color = as.factor(scores))) +
  geom_point() +
  scale_color_manual(values = c("red", "blue", "green")) +
  labs(color = "Score") +
  theme_bw()
```
